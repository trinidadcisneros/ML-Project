```python
import warnings
warnings.simplefilter('ignore')

# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Import KNeighborsClassifier from sklearn.neighbors
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.model_selection import train_test_split
```

# Read the CSV and Perform Basic Data Cleaning
The data has been subsetted to include only 100 observations to reduce wait time.

```python
df = pd.read_csv("cumulative.csv")
df = df.drop(columns=["rowid", "kepid", "kepoi_name", "kepler_name", "koi_pdisposition", "koi_score", "koi_tce_delivname", 'koi_steff', 'koi_steff_err1', 'koi_steff_err2', 'koi_slogg', 'koi_slogg_err1', 'koi_slogg_err2', 'koi_srad', 'koi_srad_err1', 'koi_srad_err2', 'ra', 'dec', 'koi_kepmag'])
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df_subset = df.head(n=100)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# Create arrays for the features and the response variable
y = df_subset["koi_disposition"].values
X = df_subset.drop('koi_disposition', axis=1).values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

# Pre-processing

Scale the data using the MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

mm_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = mm_scaler.fit(X_train)
X_test_scaled = mm_scaler.fit(X_test)
```

# Train the Support Vector Machine

```python
from sklearn.svm import SVC 
model = SVC(kernel='linear')
```

```python
model.fit(X_train, y_train)
```

```python
# Model Accuracy
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the `C` and `gamma` parameters

```python
# Train the model with GridSearch
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10], "gamma":[0.0001, 0.001, 0.01] }
grid = GridSearchCV(model, param_grid, verbose=3)
grid.fit(X, y)
```

```python
grid.best_params_
grid.best_score_
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
# Make predictions with the hypertuned model
predictions = grid.predict(X_test)
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=["CONFIRMED", "FALSE POSITIVE", "CANDIDATE"]))
```
